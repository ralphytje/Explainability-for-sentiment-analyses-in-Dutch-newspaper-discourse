```python
import os
import pandas as pd

# Define file paths for the third-labeler csv files
file_paths = {
    '1970s_coal': r"C:\Users\schol\Thesis\training model\1970s_coal.csv",
    '1970s_gas': r"C:\Users\schol\Thesis\training model\1970s_gas.csv",
    '1970s_oil': r"C:\Users\schol\Thesis\training model\1970s_oil.csv",
    '1980s_coal': r"C:\Users\schol\Thesis\training model\1980s_coal.csv",
    '1980s_gas': r"C:\Users\schol\Thesis\training model\1980s_gas.csv",
    '1980s_oil': r"C:\Users\schol\Thesis\training model\1980s_oil.csv",
    '1990s_coal': r"C:\Users\schol\Thesis\training model\1990s_coal.csv",
    '1990s_gas': r"C:\Users\schol\Thesis\training model\1990s_gas.csv",
    '1990s_oil': r"C:\Users\schol\Thesis\training model\1990s_oil.csv"
}

# Create a dictionary to store the DataFrames
df_dict = {key: pd.read_csv(path) for key, path in file_paths.items()}

# Initialize an empty dictionary to store dataframes for each year
df_per_year = {}

# Loop through each item in the df_dict
for key, df in df_dict.items():
    # Extract year and fuel type from the key
    year, fuel_type = key.split('_')

    # Add a new column for fuel type and year
    df['fuel_type'] = fuel_type
    df['year'] = year

    # Add the dataframe to the dictionary using year as key
    if year not in df_per_year:
        df_per_year[year] = df
    else:
        # If dataframe for the year already exists, merge them
        df_per_year[year] = pd.concat([df_per_year[year], df])

# Function to check if 'labeler3' values exist in 'labels_edo' or 'labels_marin'
def check_labels(data_dict):
    results = {}
    for key, df in data_dict.items():
        # Check if 'labeler3' values are in 'labels_edo' or 'labels_marin'
        try:
            exists_in_either = df['labeler3'].isin(df['labels_edo']).any() or df['labeler3'].isin(df['labels_marin']).any()
            results[key] = exists_in_either
        except:
            print(f'An exception occurred for {key}')

    return results

results_dict = check_labels(df_per_year)

for key, val in results_dict.items():
    print(key, val)

# Rename columns and select specific columns
for key, df in df_per_year.items():
    try:
        df_per_year[key] = df[['text_split', 'labeler3', 'fuel_type', 'year']].rename(columns={'text_split': 'text', 'labeler3': 'labels'})
    except KeyError:
        print(f'An exception occurred for {key}')
#%%
# Define file paths for the merged data csv files
file_paths = {
    'df_merged_1960s_coal': r"C:\Users\schol\Thesis\training model\1960s_coal_merged.csv",
    'df_merged_1960s_gas': r"C:\Users\schol\Thesis\training model\1960s_gas_merged.csv",
    'df_merged_1960s_oil': r"C:\Users\schol\Thesis\training model\1960s_oil_merged.csv",
    'df_merged_1970s_coal': r"C:\Users\schol\Thesis\training model\1970s_coal_merged.csv",
    'df_merged_1970s_gas': r"C:\Users\schol\Thesis\training model\1970s_gas_merged.csv",
    'df_merged_1970s_oil': r"C:\Users\schol\Thesis\training model\1970s_oil_merged.csv",
    'df_merged_1980s_coal': r"C:\Users\schol\Thesis\training model\1980s_coal_merged.csv",
    'df_merged_1980s_gas': r"C:\Users\schol\Thesis\training model\1980s_gas_merged.csv",
    'df_merged_1980s_oil': r"C:\Users\schol\Thesis\training model\1980s_oil_merged.csv",
    'df_merged_1990s_coal': r"C:\Users\schol\Thesis\training model\1990s_coal_merged.csv",
    'df_merged_1990s_gas': r"C:\Users\schol\Thesis\training model\1990s_gas_merged.csv",
    'df_merged_1990s_oil': r"C:\Users\schol\Thesis\training model\1990s_oil_merged.csv",
}

# Create a dictionary to store the DataFrames
df_dict = {key: pd.read_csv(path) for key, path in file_paths.items()}

# Initialize an empty dictionary to store dataframes for each year
df_per_year = {}

# Loop through each file in the dictionary
for filename, df in df_dict.items():
    # Extract year and fuel type from the filename
    year, fuel_type = filename.split('_')[2], filename.split('_')[3].split('.')[0]
    
    # Add a new column for fuel type and year
    df['fuel_type'] = fuel_type
    df['year'] = year

    # Add the dataframe to the dictionary using year as key
    if year not in df_per_year:
        df_per_year[year] = df
    else:
        # If dataframe for the year already exists, merge them
        df_per_year[year] = pd.merge(df_per_year[year], df, on=['text', 'fuel_type', 'labels', 'year'], how='outer')

# Concatenate all dataframes into one
concatenated_df = pd.concat(df_per_year.values(), ignore_index=True)
```

```python
# Initialize dictionaries to store dataframes for each decade
df_per_decade = {'1960s': None, '1970s': None, '1980s': None, '1990s': None}

# Loop through each file in the dictionary
for filename, df in df_dict.items():
    # Extract year and fuel type from the filename
    year, fuel_type = filename.split('_')[2], filename.split('_')[3].split('.')[0]
    
    # Add a new column for fuel type and year
    df['fuel_type'] = fuel_type
    df['year'] = year

    # Determine the decade and add the dataframe to the corresponding dictionary
    if '196' in year:
        if df_per_decade['1960s'] is None:
            df_per_decade['1960s'] = df
        else:
            df_per_decade['1960s'] = pd.merge(df_per_decade['1960s'], df, on=['text', 'fuel_type', 'labels', 'year'], how='outer')
    elif '197' in year:
        if df_per_decade['1970s'] is None:
            df_per_decade['1970s'] = df
        else:
            df_per_decade['1970s'] = pd.merge(df_per_decade['1970s'], df, on=['text', 'fuel_type', 'labels', 'year'], how='outer')
    elif '198' in year:
        if df_per_decade['1980s'] is None:
            df_per_decade['1980s'] = df
        else:
            df_per_decade['1980s'] = pd.merge(df_per_decade['1980s'], df, on=['text', 'fuel_type', 'labels', 'year'], how='outer')
    elif '199' in year:
        if df_per_decade['1990s'] is None:
            df_per_decade['1990s'] = df
        else:
            df_per_decade['1990s'] = pd.merge(df_per_decade['1990s'], df, on=['text', 'fuel_type', 'labels', 'year'], how='outer')

# Concatenate all dataframes for each decade into one
for decade, df_decade in df_per_decade.items():
    # Drop duplicate rows
    df_decade.drop_duplicates(inplace=True)
    
    # Save the dataframe to a CSV file
    df_decade.to_csv(f'{decade}_merged.csv', index=False)
    
# Load the CSV files into Pandas DataFrames
df_1960s = pd.read_csv('1960s_merged.csv')
df_1970s = pd.read_csv('1970s_merged.csv')
df_1980s = pd.read_csv('1980s_merged.csv')
df_1990s = pd.read_csv('1990s_merged.csv')

# Now you can use these DataFrames for analysis or visualization
df_1960s.head()
```
